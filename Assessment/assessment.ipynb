```python
import pandas as pd
import numpy as np
import datetime
import math
```

```python
def hitung_umur (tanggal_lahir) :
    hari_ini = datetime.date.hari_ini()
    umur = hari_ini.tahun - tanggal_lahir.tahun - ((hari_ini.bulan, hari_ini.hari) < (tanggal_lahir.bulan, tanggal_lahir.hari))
    return umur

def hitung_luas_lingkaran(jari_jari):
    return 3.14*jari_jari*jari_jari 

nama = input("Masukkan Nama kamu : ")
NIM = input("Masukkan NIM kamu : ")

pilih_umur = input("Apakah kamu ingin memasukkan umur atau tanggal lahir? (umur/tanggal) : ").strip().lower()

if pilih_umur == "tanggal":
    tanggal_lahir_input = input("Masukkan tanggal lahir (yyyy-mm-dd) : ")
    tanggal_lahir = datetime.datetime.strptime(tanggal_lahir_input).tanggal()
    umur = hitung_umur(tanggal_lahir)
    poin = 5
else:
    umur = int(input("Masukkan umur : "))
    poin = 2

angka_favorit = float(input("Masukkan angka favorit (untuk jari-jari lingkaran) : "))
luas_lingkaran = hitung_luas_lingkaran(angka_favorit)

print("\nOutput :)")
print(f"Nama : {nama}")
print(f"NIM : {NIM}")
print(f"Umur : {umur} tahun")
print(f"Luas lingkaran berdasarkan jari-jari {angka_favorit}: {luas_lingkaran:.2f}")
if poin > 0:
    print(f"Poin tambahan : {poin}")
```
